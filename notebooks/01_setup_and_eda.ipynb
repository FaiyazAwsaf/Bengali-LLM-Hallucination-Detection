```python
# CELL 4: Set paths
# data/ is gitignored (not in the repo). Data lives in your Drive folder:
# My Drive/Bengali-LLM-Hallucination-Detection/data/
BASE   = '/content/drive/MyDrive/Bengali-LLM-Hallucination-Detection/'
DATA   = BASE + 'data/'
MODELS = BASE + 'models/'
SUBS   = BASE + 'submissions/'

import os
os.makedirs(DATA, exist_ok=True)
os.makedirs(MODELS, exist_ok=True)
os.makedirs(SUBS, exist_ok=True)
```

```python
# CELL 5: Load data
import json
import pandas as pd

test = pd.read_csv(DATA + 'test set.csv')
sample_sub = pd.read_csv(DATA + 'sample submission.csv')

with open(DATA + 'dataset samples.json', encoding='utf-8') as f:
    samples = json.load(f)
train = pd.DataFrame(samples)

print('train (labeled samples):', train.shape)
print('test (unlabeled):', test.shape)
print('sample submission:', sample_sub.shape)
train.head()
```

```python
# CELL 6: EDA — class balance, context nulls, text lengths
print('Label distribution:')
print(train['label'].value_counts())
print()
print('NULL context ratio:', (train['context'] == '[NULL]').mean())
print()
train['response_len'] = train['response_bn'].str.len()
train['prompt_len'] = train['prompt_bn'].str.len()
print(train[['prompt_len', 'response_len']].describe())
```
